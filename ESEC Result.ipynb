```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
index = ['Type',
 'Shape',
 'Adapting Data',
 'Data Cleaning',
 'Feature Selection/Selection',
 'Model Creation',
 'Loading/Storing Model',
 'Model Selection',
 'Model Conversion',
 'Estimator Selection',
 'Error/Exception',
 'Choice of parameter',
 'Loss Function',
 'Optimizer',
 'Regularization',
 'Performance',
 'Accuracy',
 'Evaluation Strategy',
 'Visualization',
 'Output interpretation',
 'Tuning Strategy',
 'Selecting tuning parameters',
 'Prediction Accuracy',
 'Reusing pre trained model',
 'Robustness',
 'Non ML API',
 'Installation/platform/dependency',
 'Override/Add Custom Code',
 'Bug']
data = pd.read_csv("heatmapdata.csv", na_values=['#VALUE!', '#DIV/0!'])
cols = data.columns.tolist()
cols = cols[1:]
cols
```

```python
sub = data[cols]
sub = sub.fillna(0)
#sub.reindex(cols)
sub
```

```python
ax = plt.figure(figsize=(18,10))
ax= plt.subplot(111)
sns.heatmap(sub,yticklabels=cols,ax = ax)
plt.title('r (correlation coefficient among different sub categories)')
plt.figure(figsize=(18,10))
plt.show()
```

```python
rr = sub.apply(lambda x : x*x)
ax = plt.figure(figsize=(18,10))
ax= plt.subplot(111)
sns.heatmap(rr,yticklabels=cols,ax = ax,annot = False)
plt.title('r^2 (Square of Correlation coefficient) among different sub classifications')
plt.figure(figsize=(18,10))
plt.show()
```

```python
cr2 = pd.read_csv("cor2.csv")
cr2 = cr2.pivot("Stage","Library","Value")
#cr2 = np.corrcoef(cr2)
ax = plt.figure(figsize=(18,10))
ax= plt.subplot(111)
sns.heatmap(cr2,ax = ax,annot = True)
plt.title("Heatmap showing normalized percentage of questions different libraries at different stages")
plt.show()
```

```python
crdata = pd.read_csv('crr.csv')
#print(crdata)
titles = crdata.columns.tolist()
crdata = np.array(crdata).T
#print(crdata)
cr = np.corrcoef(crdata)
#print(cr)
ax = plt.figure(figsize=(18,10))
ax= plt.subplot(111)
sns.heatmap(cr,xticklabels = titles, yticklabels = titles,ax = ax,annot = True)
plt.title("Correlation between different libraries")
plt.show()
```

```python
from matplotlib import cm as cm

mask = np.zeros_like(cr)
mask[np.triu_indices_from(mask)] = True
ax = plt.figure(figsize=(18,10))
ax= plt.subplot(111)
sns.heatmap(cr,mask = mask,xticklabels = titles, yticklabels = titles,ax = ax,annot = True)
plt.title("Correlation between different libraries")
plt.show()
```
